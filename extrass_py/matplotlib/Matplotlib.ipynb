```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
##Use of function
plt.plot(x, y, "r-")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r-")

plt.subplot(1,2,2)
plt.plot(y,x,"b-")
```

```python
##objet oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.6,0.2,0.2,0.2])

axes1.plot(x,y)
axes2.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('first plot')
axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

## Figure size and DPI

```python
# fig = plt.figure(figsize=(3,2))
# ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y)
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig.savefig('mypic.png',dpi=200)
```

```python
fig = plt.figure(figsize=(3,2)) 
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')

ax.legend(loc=0)
```
